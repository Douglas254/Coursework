```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_sex = LabelEncoder()
```

```python
df_new['le_sex'] = le_sex.fit_transform(df_new['Sex'])
```

```python
df_trans=df_new.drop(['Sex'],axis='columns')
df_trans.head()
```

```python
df_trans.isnull().sum()
```

```python
target=df_trans['Survived']
```

```python
inputs=df_trans.drop(['Survived'],axis='columns')
inputs.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.2)
```

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
