# Linear Interpolation using the sin() function
This is the simplest technique where it assumes that data follows a straight line between djacent measurements

- To achieve this we use: `numpy.interp(x_values, x_points, y_points`

```python
import numpy as np
import matplotlib.pyplot as plt 

x_points = np.linspace(0, 2*np.pi, 10) # 10 equidistant x coords from 0 to 10
y_points = np.sin(x_points)

x_values = np.linspace(0,2*np.pi,50) # 50 desired points
y_values = np.interp(x_values,x_points,y_points)

plt.plot(x_points, y_points,'o') # plot known data points
plt.plot(x_values,y_values,'-x') # plot interpolated points
plt.show()
```

```python
x_points = np.linspace(0, 2*np.pi, 100) # 100 equidistant
y_points = np.sin(x_points)

x_values = np.linspace(0,2*np.pi,50) # 50 desired points
y_values = np.interp(x_values,x_points,y_points)

plt.plot(x_points, y_points,'o') # plot known data points
plt.plot(x_values,y_values,'-x') # plot interpolated points
plt.show()
```

# Cubic spline interpolation
- Just as a linear interpolation is made up of linear segments – a cubic spline
interpolation is made of segments of cubic polynomials, whose gradients
match up at the measured data points.

- These cubic polynomials are continuous up to their 2nd derivative.

- To achieve this we use: `scipy`


```python
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt

x_points = np.linspace(0, 2*np.pi, 10) # 10 equidistant x coords from 0 to 10
y_points = np.sin(x_points)

splines = interpolate.splrep(x_points, y_points)
x_values = np.linspace(0, 2*np.pi, 50) # 50 desired points
y_values = interpolate.splev(x_values, splines)

plt.plot(x_points, y_points, 'o') # plot known data points
plt.plot(x_values, y_values, '-x') # plot interpolated points
plt.show() 
```

# Curve fitting
- Collected data always contains some degree of error or imprecision

- Whereas interpolation is used when we assume that all data points are
accurate and we want to infer new intermediate data points
- curve fitting is used when we want to match an analytical (or symbolic) model
to a set of measurements which may contain some error.

## Linear regression
- One method of curve fitting is linear regression – it minimizes the `"square of
the errors"` (where the "error" is the distance each point is from the line).

## Polynomial regression
- Linear regression is a special case of polynomial regression –
since a line (i.e., an equation of the form `ax + b`) is a simple polynomial.
- But your data may not reflect a linear relationship – a polynomial of a higher
order may be a better fit.
- Both linear and non-linear polynomial regression can be done with Numpy's
polyfit function:
`numpy.polyfit(x, y, degree)`
- It returns the coeffficients for the polynomial;
the easiest way to then use these in code is to use the `numpy.poly1d` class.

```python
# Linear regression
import numpy as np

x_points = np.arange(0,6,1)
y_points = np.array([15, 10, 9, 6, 2, 0])

f = np.poly1d( np.polyfit(x_points, y_points, 1)) # linear regression
x_values = np.linspace(0, 6, 100)

plt.plot(x_points, y_points, '.')
plt.plot(x_values, f(x_values), '-')
plt.show()
```

# Polynomial regression
- If we have 6 data points, then a fifth-order polynomial will be able to give a
perfect fit for them
(i.e., there is some fifth-order polynomial on which all the data points fall
exactly).


```python
import numpy as np

x_points = np.arange(0,6,1)
y_points = np.array([15, 10, 9, 6, 2, 0])

f = np.poly1d( np.polyfit(x_points, y_points, 5)) # 5th-order polynomial
x_values = np.linspace(0, 6, 100)

plt.plot(x_points, y_points, '.')
plt.plot(x_values, f(x_values), '-')
plt.show()
```
